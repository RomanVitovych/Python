```python
no_duplicate = df.drop_duplicates(subset = 'applicant_id').copy()
no_duplicate['External Rating'] = no_duplicate['External Rating'].fillna(value = 0)
no_duplicate['Education level'] = no_duplicate['Education level'].replace(to_replace=" ", value='Середня').fillna(value='Середня')
display(no_duplicate)
```

```python
merged = pd.merge(left = df, right = df_additional, on = 'Industry', how = 'inner')
display(merged)
```

```python
merged['Raiting'] = 0
def calculated_raiting (row) :
    raiting = 0
    if pd.isna(row['Amount']) or row['External Rating'] == 0:
        return 0
    if (row['Age'] >= 35) & (row['Age'] <= 55):
        raiting += 20
    if not pd.isna(row['Applied at']):    
        raiting_date =  pd.to_datetime(row['Applied at']) 
        if raiting_date.weekday() <= 4:
            raiting += 20
    if row['Marital status'] == 'Married':
        raiting += 20
    if row['Location'] == 'Київ чи область':
        raiting += 10
    score_value = row['Score']
    if (score_value != 0) & (not pd.isna(score_value)):
        raiting += score_value
    if not pd.isna(row['External Rating']):   
        int_numb = int(row['External Rating'])
        if int_numb >= 7:
            raiting += 20
        elif int_numb <= 2:
            raiting -= 20           
    return min(raiting, 100)
merged['Raiting'] = merged.apply(calculated_raiting, axis = 1)
display(merged)
```

```python
filtered = merged[merged['Raiting'] > 0].copy()
display(filtered)
```

```python
filtered['Applied at'] = pd.to_datetime(filtered['Applied at'], format='mixed', dayfirst=True, errors='coerce')
filtered = filtered[(filtered['Applied at'].notna()) & (filtered['Raiting'].notna())].copy()
filtered.loc[:, 'Year applied'] = filtered['Applied at'].dt.isocalendar().year
filtered.loc[:, 'Week applied'] = filtered['Applied at'].dt.isocalendar().week
filtered['Year-Week'] = filtered['Year applied'].astype(str) + '-W' + filtered['Week applied'].astype(str).str.zfill(2)
#filtered.loc[:, 'Average raiting by week'] = (filtered.groupby('Week')['Raiting'].transform('mean').round(2))
#pivot_table_by_weeks = filtered.pivot_table(index='Week', values='Raiting', aggfunc='mean').round(2)
#display(pivot_table_by_weeks)
merged = filtered.groupby('Year-Week').agg(AverageRaiting=('Raiting', 'mean')).reset_index().round(2)
display(merged)
```
