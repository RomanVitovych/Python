```python
df_main['LanguageHaveWorkedWith'] = df_main['LanguageHaveWorkedWith'].astype(str)
worked_with_Python = df_main['LanguageHaveWorkedWith'].str.contains('Python', case=False, na=False).sum()
count_all_responses = df_main['LanguageHaveWorkedWith'].str.len().gt(0).sum()
percentage_in_python = round(worked_with_Python * 100 / count_all_responses, 2) 
display(percentage_in_python)
```

```python
learn_online = df_main['LearnCode'].str.contains('online courses', case=False, na=False).sum()
print(learn_online)
```

```python
df_worked_with_python = df_main[
    (df_main['LanguageHaveWorkedWith'].str.contains('Python', case=False, na=False)) &
    (df_main['ConvertedCompYearly'].notna())
]
compensation_by_countries = (
    df_worked_with_python
    .groupby('Country', as_index=False)
    .agg(
        AverageCompensation=('ConvertedCompYearly', 'mean'),
        MedianCompensation=('ConvertedCompYearly', 'median')
    )
    .round(2)
)
display(compensation_by_countries)
```

```python
compensation = df_main.sort_values('ConvertedCompYearly', ascending = False).head(5)
print(compensation[['ConvertedCompYearly','EdLevel']])
```

```python
df_main['UsesPython'] = df_main['LanguageHaveWorkedWith'].str.contains('Python', na = False)
result = df_main.groupby('Age')['UsesPython'].agg('mean') * 100
print(round(result, 2))
```

```python
remote_df = df_main[(df_main['RemoteWork'] == 'Remote') & (df_main['ConvertedCompYearly'].notna())]
percentile_75_remote = remote_df['ConvertedCompYearly'].dropna().quantile(0.75)
high_percentile_remote_df = remote_df[remote_df['ConvertedCompYearly'] >= percentile_75_remote]
most_popular_industries = high_percentile_remote_df['Industry'].dropna().value_counts()
filtered_industries = most_popular_industries.drop(index='Other:', errors='ignore')
top_5_industries = filtered_industries.head(5).reset_index()
top_5_industries.index+=1
print("Найпоширеніші галузі серед віддалених працівників з високою компенсацією:\n" f"{top_5_industries}")
```
